```python
# ============================================
# EMI PROJECT - EXPLORATORY DATA ANALYSIS
# ============================================

# =========================
# 1. IMPORT LIBRARIES
# =========================
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10

sns.set_style("whitegrid")

# =========================
# 2. LOAD DATA
# =========================
DATA_PATH = r"../data/processed/emi_featured_final.csv"

df = pd.read_csv(DATA_PATH)

print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nFirst 5 Rows:")
display(df.head())

# =========================
# 3. BASIC DATA OVERVIEW
# =========================
print("\nData Types:\n")
print(df.dtypes)

print("\nMissing Values:\n")
print(df.isnull().sum().sort_values(ascending=False))

print("\nDuplicate Rows:", df.duplicated().sum())

# =========================
# 4. STATISTICAL SUMMARY
# =========================
print("\nNumerical Summary:\n")
display(df.describe().T)

print("\nCategorical Summary:\n")
display(df.describe(include="object").T)

# =========================
# 5. TARGET VARIABLE CHECK
# =========================
print("\nEMI Eligibility Distribution:")
print(df["emi_eligibility"].value_counts())
print("\nEMI Eligibility Proportion:")
print(df["emi_eligibility"].value_counts(normalize=True) * 100)

# =========================
# 6. HELPER FUNCTION
# =========================
def add_percentage_labels(ax, total=None):
    """
    Add percentage labels on countplot bars.
    """
    if total is None:
        total = sum([p.get_height() for p in ax.patches])

    for p in ax.patches:
        height = p.get_height()
        percentage = 100 * height / total if total > 0 else 0
        ax.annotate(f"{percentage:.1f}%",
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=10)

# =========================
# 7. EMI ELIGIBILITY DISTRIBUTION
# =========================
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x="emi_eligibility")
plt.title("EMI Eligibility Distribution")
plt.xlabel("EMI Eligibility")
plt.ylabel("Count")
add_percentage_labels(ax)
plt.show()

# Insight Placeholder:
print("Insight: Check whether the dataset is balanced or imbalanced in terms of EMI approval.")

# =========================
# 8. EMI ELIGIBILITY ACROSS LENDING SCENARIOS
# =========================
if "emi_scenario" in df.columns:
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=df, x="emi_scenario", hue="emi_eligibility")
    plt.title("EMI Eligibility Across Lending Scenarios")
    plt.xlabel("EMI Scenario")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.legend(title="EMI Eligibility")
    plt.show()

    scenario_summary = pd.crosstab(df["emi_scenario"], df["emi_eligibility"], normalize="index") * 100
    print("\nEMI Eligibility % within each lending scenario:")
    display(scenario_summary)

    print("Insight: Compare approval rates across lending scenarios to identify strict vs favorable cases.")

# =========================
# 9. FINANCIAL VARIABLES VS EMI ELIGIBILITY
# =========================
financial_cols = [
    "monthly_salary",
    "credit_score",
    "bank_balance",
    "emergency_fund",
    "current_emi_amount",
    "requested_amount",
    "max_monthly_emi",
    "total_expenses",
    "disposable_income",
    "expense_to_income_ratio",
    "emi_to_income_ratio",
    "loan_to_income_ratio",
    "disposable_income_ratio",
    "requested_emi_to_income_ratio",
    "emi_capacity_gap"
]

available_financial_cols = [col for col in financial_cols if col in df.columns]

for col in available_financial_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x="emi_eligibility", y=col)
    plt.title(f"{col} vs EMI Eligibility")
    plt.xlabel("EMI Eligibility")
    plt.ylabel(col)
    plt.show()

    print(f"Insight for {col}: Compare the distribution of {col} between eligible and non-eligible applicants.")

# =========================
# 10. CORRELATION HEATMAP
# =========================
corr_cols = [
    "monthly_salary",
    "years_of_employment",
    "monthly_rent",
    "dependents",
    "school_fees",
    "college_fees",
    "travel_expenses",
    "groceries_utilities",
    "other_monthly_expenses",
    "current_emi_amount",
    "credit_score",
    "bank_balance",
    "emergency_fund",
    "requested_amount",
    "requested_tenure",
    "max_monthly_emi",
    "total_expenses",
    "disposable_income",
    "expense_to_income_ratio",
    "emi_to_income_ratio",
    "savings_to_income_ratio",
    "loan_to_income_ratio",
    "emergency_fund_months",
    "disposable_income_ratio",
    "emi_capacity_gap",
    "requested_emi_to_income_ratio",
    "emi_eligibility"
]

available_corr_cols = [col for col in corr_cols if col in df.columns]

corr_df = df[available_corr_cols].copy()

# Convert target if needed
if corr_df["emi_eligibility"].dtype == "object":
    corr_df["emi_eligibility"] = corr_df["emi_eligibility"].map({"Eligible": 1, "Not Eligible": 0, "Yes": 1, "No": 0})

plt.figure(figsize=(16, 10))
sns.heatmap(corr_df.corr(), annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Financial Variables")
plt.show()

# Correlation with targets
print("\nCorrelation with EMI Eligibility:")
display(corr_df.corr()["emi_eligibility"].sort_values(ascending=False))

if "max_monthly_emi" in corr_df.columns:
    print("\nCorrelation with Maximum Monthly EMI:")
    display(corr_df.corr()["max_monthly_emi"].sort_values(ascending=False))

print("Insight: Identify which variables are most positively and negatively associated with approval and EMI capacity.")

# =========================
# 11. DEMOGRAPHIC PATTERNS
# =========================
demographic_cols = [
    "gender",
    "marital_status",
    "education",
    "employment_type",
    "company_type",
    "house_type"
]

available_demographic_cols = [col for col in demographic_cols if col in df.columns]

for col in available_demographic_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, hue="emi_eligibility")
    plt.title(f"EMI Eligibility by {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.legend(title="EMI Eligibility")
    plt.show()

    summary = pd.crosstab(df[col], df["emi_eligibility"], normalize="index") * 100
    print(f"\nApproval % by {col}:")
    display(summary)

    print(f"Insight for {col}: Check whether certain applicant groups show consistently higher or lower EMI approval rates.")

# =========================
# 12. AGE / EXPERIENCE / FAMILY RELATIONSHIPS
# =========================
extra_demo_cols = [
    "age",
    "family_size",
    "dependents",
    "years_of_employment",
    "experience_bucket",
    "salary_bucket"
]

available_extra_demo_cols = [col for col in extra_demo_cols if col in df.columns]

for col in available_extra_demo_cols:
    if df[col].dtype == "object":
        plt.figure(figsize=(10, 5))
        sns.countplot(data=df, x=col, hue="emi_eligibility")
        plt.title(f"EMI Eligibility by {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.legend(title="EMI Eligibility")
        plt.show()

        summary = pd.crosstab(df[col], df["emi_eligibility"], normalize="index") * 100
        print(f"\nApproval % by {col}:")
        display(summary)

    else:
        plt.figure(figsize=(8, 5))
        sns.boxplot(data=df, x="emi_eligibility", y=col)
        plt.title(f"{col} vs EMI Eligibility")
        plt.xlabel("EMI Eligibility")
        plt.ylabel(col)
        plt.show()

    print(f"Insight for {col}: Observe whether this demographic or experience variable influences EMI approval.")

# =========================
# 13. RISK FACTOR RELATIONSHIPS
# =========================
risk_cols = [
    "high_financial_stress",
    "low_emergency_fund_flag",
    "high_loan_burden_flag",
    "credit_score_bucket"
]

available_risk_cols = [col for col in risk_cols if col in df.columns]

for col in available_risk_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, hue="emi_eligibility")
    plt.title(f"EMI Eligibility by {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.legend(title="EMI Eligibility")
    plt.show()

    summary = pd.crosstab(df[col], df["emi_eligibility"], normalize="index") * 100
    print(f"\nApproval % by {col}:")
    display(summary)

    print(f"Insight for {col}: Evaluate whether this engineered risk feature separates eligible and non-eligible applicants.")

# =========================
# 14. AFFORDABILITY & REPAYMENT CAPACITY
# =========================
scatter_pairs = [
    ("monthly_salary", "max_monthly_emi"),
    ("disposable_income", "max_monthly_emi"),
    ("credit_score", "max_monthly_emi"),
    ("requested_amount", "max_monthly_emi"),
    ("emi_to_income_ratio", "max_monthly_emi"),
    ("loan_to_income_ratio", "max_monthly_emi"),
]

for x_col, y_col in scatter_pairs:
    if x_col in df.columns and y_col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(data=df.sample(min(5000, len(df)), random_state=42),
                        x=x_col, y=y_col, hue="emi_eligibility", alpha=0.6)
        plt.title(f"{y_col} vs {x_col}")
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.show()

        print(f"Insight for {x_col} vs {y_col}: Analyze how repayment capacity changes with {x_col}.")

# =========================
# 15. BUSINESS SUMMARY TABLES
# =========================
print("\nAverage Financial Metrics by EMI Eligibility:\n")
business_cols = [
    "monthly_salary",
    "credit_score",
    "bank_balance",
    "emergency_fund",
    "requested_amount",
    "max_monthly_emi",
    "total_expenses",
    "disposable_income",
    "expense_to_income_ratio",
    "emi_to_income_ratio",
    "loan_to_income_ratio",
    "disposable_income_ratio",
    "emi_capacity_gap"
]
available_business_cols = [col for col in business_cols if col in df.columns]

display(df.groupby("emi_eligibility")[available_business_cols].mean().T)

print("\nMedian Financial Metrics by EMI Eligibility:\n")
display(df.groupby("emi_eligibility")[available_business_cols].median().T)

# =========================
# 16. OPTIONAL - APPROVAL RATE BY BINS
# =========================
if "monthly_salary" in df.columns:
    df["salary_band_eda"] = pd.qcut(df["monthly_salary"], q=5, duplicates="drop")
    salary_band_summary = pd.crosstab(df["salary_band_eda"], df["emi_eligibility"], normalize="index") * 100

    plt.figure(figsize=(10, 5))
    salary_band_summary.plot(kind="bar", stacked=True)
    plt.title("EMI Eligibility by Salary Band")
    plt.xlabel("Salary Band")
    plt.ylabel("Percentage")
    plt.xticks(rotation=45)
    plt.legend(title="EMI Eligibility")
    plt.show()

    print("\nApproval % by Salary Band:")
    display(salary_band_summary)

if "credit_score" in df.columns:
    df["credit_band_eda"] = pd.qcut(df["credit_score"], q=5, duplicates="drop")
    credit_band_summary = pd.crosstab(df["credit_band_eda"], df["emi_eligibility"], normalize="index") * 100

    plt.figure(figsize=(10, 5))
    credit_band_summary.plot(kind="bar", stacked=True)
    plt.title("EMI Eligibility by Credit Score Band")
    plt.xlabel("Credit Score Band")
    plt.ylabel("Percentage")
    plt.xticks(rotation=45)
    plt.legend(title="EMI Eligibility")
    plt.show()

    print("\nApproval % by Credit Score Band:")
    display(credit_band_summary)


# =========================
# 18. SAVE OPTIONAL EDA SUMMARY
# =========================
eda_summary = {
    "dataset_shape": df.shape,
    "numerical_columns_count": len(df.select_dtypes(include=np.number).columns),
    "categorical_columns_count": len(df.select_dtypes(include="object").columns),
    "emi_eligibility_distribution": df["emi_eligibility"].value_counts(dropna=False).to_dict()
}

summary_path = r"../data/processed/eda_summary.json"
pd.Series(eda_summary).to_json(summary_path)

print(f"\nEDA summary saved to: {summary_path}")
```

```python
# =========================
# STEP 4: CORRELATION ANALYSIS
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
DATA_PATH = r"../data/processed/emi_featured_final.csv"
df = pd.read_csv(DATA_PATH)

# Encode target for correlation
df["emi_eligibility_encoded"] = df["emi_eligibility"].map({
    "Not_Eligible": 0,
    "Eligible": 1,
    "High_Risk": 2
})

# Select numeric columns
num_df = df.select_dtypes(include=["number"])

# Correlation matrix
corr = num_df.corr()

# Heatmap
plt.figure(figsize=(14,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()

# Correlation with target
print("\nCorrelation with EMI Eligibility:\n")
print(corr["emi_eligibility_encoded"].sort_values(ascending=False))
```

Correlation analysis shows that max_monthly_emi and emi_capacity_gap have the strongest positive relationship with EMI eligibility, indicating that higher repayment capacity significantly improves approval chances.

Financial strength indicators such as disposable income, bank balance, and monthly salary also positively influence eligibility, confirming that financially stable applicants are more likely to be approved.

On the other hand, risk indicators such as high_financial_stress, high_loan_burden_flag, and high requested EMI ratios show strong negative correlation with eligibility. This suggests that higher financial burden and repayment pressure reduce the likelihood of loan approval.

Overall, the results highlight that EMI eligibility is primarily driven by repayment capacity and financial stability, while financial stress and excessive loan burden act as major rejection factors.

```python
# =========================
# STEP 5: DEMOGRAPHIC ANALYSIS
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
DATA_PATH = r"../data/processed/emi_featured_final.csv"
df = pd.read_csv(DATA_PATH)

cols = [
    "gender",
    "marital_status",
    "education",
    "employment_type",
    "house_type"
]

for col in cols:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col, hue="emi_eligibility")
    plt.title(f"EMI Eligibility by {col}")
    plt.xticks(rotation=45)
    plt.show()

    summary = pd.crosstab(df[col], df["emi_eligibility"], normalize="index") * 100
    print(f"\n{col} Summary:\n")
    print(summary)
```

```python
# =========================
# STEP 6: RISK FEATURE ANALYSIS
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
DATA_PATH = r"../data/processed/emi_featured_final.csv"
df = pd.read_csv(DATA_PATH)

cols = [
    "high_financial_stress",
    "low_emergency_fund_flag",
    "high_loan_burden_flag",
    "credit_score_bucket"
]

for col in cols:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col, hue="emi_eligibility")
    plt.title(f"EMI Eligibility by {col}")
    plt.xticks(rotation=45)
    plt.show()

    summary = pd.crosstab(df[col], df["emi_eligibility"], normalize="index") * 100
    print(f"\n{col} Summary:\n")
    print(summary)
```

Risk feature analysis shows that engineered features play a critical role in distinguishing between eligible and non-eligible applicants.

Among all features, high_financial_stress emerges as the strongest predictor, with applicants under financial stress showing extremely low approval rates and a high likelihood of rejection.

Similarly, high_loan_burden_flag demonstrates a strong negative impact on eligibility, where applicants with high loan burdens are almost always rejected.

Low emergency fund levels also reduce approval probability, indicating the importance of financial safety buffers.

While credit score remains an important factor, the engineered risk features provide better separation between classes, highlighting the effectiveness of feature engineering in improving model performance.
